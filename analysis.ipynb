```python
nan_rows = market_cap_data.isna().any(axis=1)
market_cap_data[nan_rows]
```

```python
market_cap_data.dropna(axis=0, inplace=True)
```

```python
indices_to_be_dropped = market_cap_data[market_cap_data['MarketCap_upto_31/03/2023_(in Crores)']=='*Not available for trading as on March 31, 2023'].index
market_cap_data.drop(indices_to_be_dropped, inplace=True)
```

Conditional dropping of indices/rows; Much cleaner than simply hardcoding the index values of the rows to be dropped

```python
#((market_cap_data['MarketCap_upto_31/03/2023'].astype(float)/100).round()).astype(int)
```

```python
market_cap_data['MarketCap_upto_31/03/2023_(in Crores)'] = ((market_cap_data['MarketCap_upto_31/03/2023_(in Crores)'].astype(float)/100).round()).astype(int)
```

We convert the series dtype to 'float' so that we can divide it by 100(conversion to Crores) and then round it up using the rounf method. Then finally we convert the entire dtype of the series to 'int' so that we get rid of the .0 decimal.
